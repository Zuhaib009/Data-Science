```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step 1: Load the dataset
healthcare =  pd.read_csv('healthcare.csv')
healthcare.head()
```

```python
# Let’s grab only the rows in the dataset that are about chest pain.
chest_pain = healthcare[healthcare['DRG Definition'] == '313 - CHEST PAIN']
chest_pain
```

```python
# get every chest pain diagnosis in Alabama
alabama_chest_pain = chest_pain[chest_pain['Provider State'] == "AL"]
alabama_chest_pain
```

```python
# find the average cost of those diagnoses
costs = alabama_chest_pain[' Average Covered Charges '].values
costs
```

```python
# Create a box plot
plt.boxplot(costs)

# Set the plot title and y-axis label
plt.title('Box Plot')
plt.ylabel('Costs')

# Display the plot
plt.show()
```

```python
# We’ll now use a for loop to separate the dataset into a dataset for each state:
datasets = []
for state in chest_pain['Provider State'].unique():
    datasets.append(chest_pain[chest_pain['Provider State'] == state][' Average Covered Charges '].values)
    
datasets
```

```python
# Create a box plot
plt.figure(figsize=(20,6))
plt.boxplot(datasets)

# Set the plot title and y-axis label
plt.title('Box Plot')
plt.ylabel('States')

# Display the plot
plt.show()
```
